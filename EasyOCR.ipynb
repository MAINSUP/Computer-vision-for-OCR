## Testing easyOCR lib on handwriting text in an image

#### Importing text prediction and visualization libraries

```python
import easyocr
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\msste\OneDrive\Coding\DA\Meduzzen CV internship\OCR\test image.jpg")
plt.imshow(img)
```

### Performing text detection

```python
reader = easyocr.Reader(['en']) # specify the language
```

Getting results with prediction score

```python
result = reader.readtext(img)
for (bbox, text, prob) in result:
    print(f'Text: {text}, Probability: {prob}')
```

As can be seen handwriting recognition is not good

### Converting to gray scale for comparison test

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
result2 = reader.readtext(gray)
for (bbox, text, prob) in result2:
    print(f'Text: {text}, Probability: {prob}')
```

Changing color scheme did not improve any performance

```python
img2 = cv2.imread(r"C:\Users\msste\OneDrive\Coding\DA\Meduzzen CV internship\OCR\test image 2.jpg")
plt.imshow(img2)
```
